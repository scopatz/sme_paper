```python
import re
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc('font', family='serif', size=16)

rncomp = re.compile('generating (\d+) component enrichment')
rgentimes = re.compile('completed in (\S*?) s')
rops = re.compile('reduced (\d+) ops to (\d+)')

ragged_ave = lambda xrag: [sum(x)/len(x) for x in xrag]

ncomps = range(3, 41)
gentimes = [[] for n in range(38)]
fullops = [[] for n in range(38)]
cseops = [[] for n in range(38)]

logfiles = ['sme05.log', 'sme10.log', 'sme15.log', 'sme20.log', 'sme30.log', 'sme40.log', ]
for logfile in logfiles:
    with open(logfile) as f:
        log = f.read()
    mncomps = [int(m.group(1))-3 for m in rncomp.finditer(log)]
    mgentimes = [float(m.group(1)) for m in rgentimes.finditer(log)]
    mfullops = [float(m.group(1)) for m in rops.finditer(log)]
    mcseops = [float(m.group(2)) for m in rops.finditer(log)]
    for i, j in enumerate(mncomps):
        gentimes[j].append(mgentimes[i])
        fullops[j].append(mfullops[i])
        cseops[j].append(mcseops[i])
gentimes = ragged_ave(gentimes)
fullops = ragged_ave(fullops)
cseops = ragged_ave(cseops)
```

```python
fig = figure(figsize=(7,7))
plt.plot(ncomps, gentimes, 'k-', figure=fig)
plt.xlabel('number of components')
plt.ylabel('code generation time [s]')
plt.savefig('../codegen_times.eps', bbox_inches="tight")
```

```python
fig = figure(figsize=(7,7))
plt.semilogy(ncomps, cseops, 'k-', label="post-CSE", figure=fig)
plt.semilogy(ncomps, fullops, 'k--', label="pre-CSE",figure=fig)
plt.legend(loc=0)
plt.xlabel('number of components')
plt.ylabel('number of operations')
plt.savefig('../numops.eps', bbox_inches="tight")
```
